# Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
clean_data = pd.read_csv("Downloads/pds-assignment-master/Q2/data_clean/clean_data.csv")

```

```python
# Scatter Plot: Math vs Reading Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', hue='gender', data=clean_data, palette='Set2')
plt.title('Math vs Reading Scores')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Math vs Reading Scores.png")
plt.show()
```

```python
# Bar Plot: Average Math Score by Parental Education
plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='math score', data=clean_data, palette='pastel')
plt.title('Average Math Score by Parental Education')
plt.xticks(rotation=45, ha='right')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Average Math Score by parental Education.png")
plt.show()
```

```python
# Grouped Bar Plot: Average Scores by Test Preparation Course and Gender
plt.figure(figsize=(10, 6))
sns.barplot(x='test preparation course', y='math score', hue='gender', data=clean_data, palette='Set3')
plt.title('Average Math Score by Test Preparation Course and Gender')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Average Math Score by Test Preparation Course and Gender.png")
plt.show()
```

```python
# Pie Chart: Distribution of Gender
gender_counts = clean_data['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue'])
plt.title('Distribution of Gender')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Distribution of Gender")
plt.show()
```

```python
# Pair Plot: Visualize relationships between numeric variables
sns.pairplot(clean_data, hue='gender', palette='Set2')
plt.suptitle('Pair Plot of Numeric Variables by Gender', y=1.02)
plt.savefig("Downloads/pds-assignment-master/Q2/results/Pair Plot of Numeric Variables by Gender")
plt.show()
```

```python
# Box Plot: Writing Scores by Lunch Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='lunch', y='writing score', data=clean_data, palette='pastel')
plt.title('Writing Scores by Lunch Type')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Writing Scores by Lunch Type")
plt.show()
```

```python
# Bar Plot: Average Scores by Parental Education and Gender
plt.figure(figsize=(12, 8))
sns.barplot(x='parental level of education', y='math score', data=clean_data, hue='gender', palette='Set2')
plt.title('Average Math Scores by Parental Education and Gender')
plt.xticks(rotation=45, ha='right')
plt.savefig("Downloads/pds-assignment-master/Q2/results/Average Math Scores by Parental Education and Gender")
plt.show()
```
